## Introducción

El dataset del Titanic contiene información sobre los pasajeros del famoso
barco que naufragó en 1912. El objetivo de este análisis es explorar los
datos, entender las variables y responder preguntas sobre los pasajeros
y sus probabilidades de supervivencia.

Los datos provienen de Kaggle y están divididos en dos conjuntos:
- **train.csv**: 891 pasajeros con información de supervivencia

## 4. Nueva variable: Familiares

```python
df_train['Familiares'] = df_train['SibSp'] + df_train['Parch']
df_test['Familiares'] = df_test['SibSp'] + df_test['Parch']

print("Variable Familiares creada correctamente")
print(f"\nDistribucion en train:")
print(df_train['Familiares'].value_counts().sort_index())
```

### Análisis de la variable Familiares

La variable **Familiares** representa el número total de familiares
que viajaban con cada pasajero a bordo del Titanic.

**Observaciones:**
- La mayoría de los pasajeros viajaban solos (Familiares = 0)
- Los pasajeros con familias pequeñas (1 a 3 familiares) son el
  segundo grupo más grande
- Muy pocos pasajeros viajaban con familias grandes (4 o más)
- Esta variable será útil para analizar si viajar acompañado
  influyó en las probabilidades de supervivencia

## 5. Unión vertical de los datasets

```python
df_train['source'] = 'train'
df_test['source'] = 'test'

df_test['Survived'] = None

print("Columnas en train:", df_train.columns.tolist())
print("Columnas en test: ", df_test.columns.tolist())
```

```python
df_total = pd.concat([df_train, df_test], ignore_index=True)

print("Dataset combinado creado correctamente")
print(f"Filas totales: {df_total.shape[0]}")
print(f"Columnas:      {df_total.shape[1]}")
print(f"\nDistribucion por fuente:")
print(df_total['source'].value_counts())
```

### Decisión sobre la columna Survived en test

El dataset **test** no contiene la variable `Survived` porque
en la competencia de Kaggle es la variable que se debe predecir.

Para poder unir ambos datasets verticalmente sin perder información,
agregamos la columna `Survived` al dataset test con valores `NaN`,
lo que indica que no se conoce la supervivencia de esos pasajeros.

El dataset combinado tiene **1309 pasajeros** en total:
- 891 de train (con información de supervivencia)
- 418 de test (sin información de supervivencia)
