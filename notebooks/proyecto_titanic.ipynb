## 5. Unión vertical de los datasets

```python
df_train['source'] = 'train'
df_test['source'] = 'test'

df_test['Survived'] = None

print("Columnas en train:", df_train.columns.tolist())
print("Columnas en test: ", df_test.columns.tolist())
```

```python
df_total = pd.concat([df_train, df_test], ignore_index=True)

print("Dataset combinado creado correctamente")
print(f"Filas totales: {df_total.shape[0]}")
print(f"Columnas:      {df_total.shape[1]}")
print(f"\nDistribucion por fuente:")
print(df_total['source'].value_counts())
```

### Decisión sobre la columna Survived en test

El dataset **test** no contiene la variable `Survived` porque
en la competencia de Kaggle es la variable que se debe predecir.

Para poder unir ambos datasets verticalmente sin perder información,
agregamos la columna `Survived` al dataset test con valores `NaN`,
lo que indica que no se conoce la supervivencia de esos pasajeros.

El dataset combinado tiene **1309 pasajeros** en total:
- 891 de train (con información de supervivencia)
- 418 de test (sin información de supervivencia)

## 8. Preguntas concretas sobre los pasajeros

### 8a. Cual es la edad promedio de los pasajeros

```python
 # Edad promedio de los pasajeros en el dataset combinado
edad_promedio = df_total['Age'].mean()
edad_mediana = df_total['Age'].median()
edad_min = df_total['Age'].min()
edad_max = df_total['Age'].max()
faltantes_edad = df_total['Age'].isnull().sum()

print("=== Edad de los pasajeros ===")
print(f"Edad promedio:  {edad_promedio:.2f} anos")
print(f"Edad mediana:   {edad_mediana:.2f} anos")
print(f"Edad minima:    {edad_min:.0f} anos")
print(f"Edad maxima:    {edad_max:.0f} anos")
print(f"Valores faltantes en Age: {faltantes_edad}")
```

### Observacion

El promedio y la mediana de edad son similares, lo que indica una distribucion
relativamente simetrica. Sin embargo, hay valores faltantes en Age que se deben
tener en cuenta al interpretar estos resultados.

### 8c. Tarifa promedio de pasajeros de primera clase

```python
# Tarifa promedio por clase de pasajero
tarifa_por_clase = df_total.groupby('Pclass')['Fare'].mean().round(2)

print("=== Tarifa promedio por clase ===")
for clase, tarifa in tarifa_por_clase.items():
    print(f"Clase {clase}: ${tarifa}")

tarifa_primera = df_total[df_total['Pclass'] == 1]['Fare'].mean()
print(f"\nLa tarifa promedio de primera clase fue ${tarifa_primera:.2f}")
print(f"Es {(tarifa_primera / df_total[df_total['Pclass'] == 3]['Fare'].mean()):.1f} veces mas cara que tercera clase")
```

### Observacion

La tarifa de primera clase es significativamente mas alta que la de segunda
y tercera clase, lo que refleja las grandes diferencias economicas entre los
grupos de pasajeros a bordo del Titanic.

### 8e. Edad mas joven y mas vieja

```python
# Pasajero mas joven y mas viejo
edad_min = df_total['Age'].min()
edad_max = df_total['Age'].max()

pasajero_joven = df_total[df_total['Age'] == edad_min][['Name', 'Age', 'Pclass', 'Sex', 'source']].head(1)
pasajero_viejo = df_total[df_total['Age'] == edad_max][['Name', 'Age', 'Pclass', 'Sex', 'source']].head(1)

print("=== Pasajero mas joven ===")
print(pasajero_joven.to_string(index=False))

print("\n=== Pasajero mas viejo ===")
print(pasajero_viejo.to_string(index=False))
```

### 8f. Cuantos pasajeros viajaron desde cada puerto

```python
# Distribucion de pasajeros por puerto de embarque
puertos = df_total['Embarked'].value_counts()
puertos_pct = (df_total['Embarked'].value_counts() / df_total['Embarked'].count() * 100).round(2)

print("=== Pasajeros por puerto de embarque ===")
print(f"Southampton (S): {puertos.get('S', 0)} pasajeros ({puertos_pct.get('S', 0)}%)")
print(f"Cherbourg   (C): {puertos.get('C', 0)} pasajeros ({puertos_pct.get('C', 0)}%)")
print(f"Queenstown  (Q): {puertos.get('Q', 0)} pasajeros ({puertos_pct.get('Q', 0)}%)")
print(f"\nValores faltantes en Embarked: {df_total['Embarked'].isnull().sum()}")
```

### Observacion

La gran mayoria de los pasajeros embarco en Southampton, que era el puerto
principal de salida del Titanic. Cherbourg y Queenstown fueron paradas

### 8g. Cuantos pasajeros viajaron solos y cuantos con familiares

```python
# Verificamos si Familiares existe en df_total, si no la creamos
if 'Familiares' not in df_total.columns:
    df_total['Familiares'] = df_total['SibSp'] + df_total['Parch']

# Clasificacion de pasajeros segun si viajaban solos o acompanados
solos = df_total[df_total['Familiares'] == 0].shape[0]
acompanados = df_total[df_total['Familiares'] > 0].shape[0]
total = df_total.shape[0]

print("=== Pasajeros solos vs acompanados ===")
print(f"Viajaban solos:        {solos} pasajeros ({solos/total*100:.2f}%)")
print(f"Viajaban acompanados:  {acompanados} pasajeros ({acompanados/total*100:.2f}%)")
print(f"Total:                 {total} pasajeros")
```

### 9a. Sobrevivieron mas mujeres que hombres proporcionalmente

Creemos que si, las mujeres tuvieron mayor tasa de supervivencia que
los hombres porque en la evacuacion del Titanic se priorizo primero
a mujeres y ninos.

```python
# Tasa de supervivencia por sexo usando df_train
supervivencia_sexo = df_train.groupby('Sex')['Survived'].agg(['sum', 'count', 'mean'])
supervivencia_sexo.columns = ['Sobrevivieron', 'Total', 'Tasa']
supervivencia_sexo['Tasa'] = (supervivencia_sexo['Tasa'] * 100).round(2)

print("=== Supervivencia por sexo ===")
print(supervivencia_sexo.to_string())
```

### Observacion

Los datos confirman que las mujeres tuvieron una tasa de supervivencia
significativamente mayor que los hombres. Esto es consistente con la
politica de evacuacion del Titanic que priorizaba mujeres y ninos.

### 9c. Pasajeros mayores de 50 o menores de 10 sobrevivieron mas proporcionalmente

Creemos que los menores de 10 si tuvieron mayor tasa de supervivencia
por la politica de evacuacion. Sin embargo, los mayores de 50 posiblemente
no, ya que la edad avanzada podia dificultar el proceso de evacuacion.

```python
# Comparacion de supervivencia por grupo etario extremo vs resto
grupo_extremo = df_train[(df_train['Age'] < 10) | (df_train['Age'] > 50)]
grupo_resto = df_train[(df_train['Age'] >= 10) & (df_train['Age'] <= 50)]

tasa_extremo = grupo_extremo['Survived'].mean() * 100
tasa_resto = grupo_resto['Survived'].mean() * 100

print("=== Supervivencia por grupo etario ===")
print(f"Menores de 10 o mayores de 50:")
print(f"  Total:         {len(grupo_extremo)}")
print(f"  Sobrevivieron: {grupo_extremo['Survived'].sum()}")
print(f"  Tasa:          {tasa_extremo:.2f}%")

print(f"\nResto de pasajeros (10 a 50 anos):")
print(f"  Total:         {len(grupo_resto)}")
print(f"  Sobrevivieron: {grupo_resto['Survived'].sum()}")
print(f"  Tasa:          {tasa_resto:.2f}%")
```

### Observacion

Los datos muestran que los pasajeros en edades extremas tuvieron una
tasa de supervivencia ligeramente diferente al resto. Los menores de
10 anos se beneficiaron de la politica de evacuacion, mientras que
los mayores de 50 anos enfrentaron mayores dificultades fisicas
durante la evacuacion.

### 9e. Pasajeros de primera clase sobrevivieron mas que otros proporcionalmente

Creemos que si, los pasajeros de primera clase tuvieron mayor tasa de
supervivencia porque tenian cabinas en las cubiertas superiores, acceso
prioritario a los botes salvavidas y mayor capacidad economica para
obtener mejores ubicaciones en el barco.

```python
# Tasa de supervivencia por clase de pasajero
supervivencia_clase = df_train.groupby('Pclass')['Survived'].agg(['sum', 'count', 'mean'])
supervivencia_clase.columns = ['Sobrevivieron', 'Total', 'Tasa']
supervivencia_clase['Tasa'] = (supervivencia_clase['Tasa'] * 100).round(2)

print("=== Supervivencia por clase ===")
print(supervivencia_clase.to_string())
```

### Observacion

Los datos confirman que los pasajeros de primera clase tuvieron una
tasa de supervivencia significativamente mayor que los de segunda y
tercera clase. Esto refleja las desigualdades sociales de la epoca
y el acceso diferenciado a los recursos de evacuacion.

## 11. Proporcion por sexo grupo etario cabina y supervivencia

```python
# Definimos la variable grupo_etario
def clasificar_edad(age):
    if age < 10:
        return 'Nino'
    elif age < 50:
        return 'Adulto'
    else:
        return 'Mayor'

df_train['grupo_etario'] = df_train['Age'].apply(clasificar_edad)
df_train['tiene_cabina'] = df_train['Cabin'].notna()

# Tabla de proporcion por sexo, grupo etario, cabina y supervivencia
tabla = df_train.groupby(['Sex', 'grupo_etario', 'tiene_cabina', 'Survived']).size().reset_index(name='Cantidad')
tabla_pivot = df_train.groupby(['Sex', 'grupo_etario', 'tiene_cabina'])['Survived'].agg(['sum', 'count', 'mean'])
tabla_pivot.columns = ['Sobrevivieron', 'Total', 'Tasa']
tabla_pivot['Tasa'] = (tabla_pivot['Tasa'] * 100).round(2)

print("=== Proporcion de supervivencia por sexo grupo etario y cabina ===")
print(tabla_pivot.to_string())
```

### Conclusion del analisis

Las mujeres adultas con cabina asignada tuvieron la mayor tasa de
supervivencia, lo que confirma que la clase social, el sexo y la
ubicacion en el barco fueron factores determinantes. Los hombres
adultos sin cabina tuvieron la menor tasa de supervivencia, reflejando
las desigualdades en el acceso a los botes salvavidas durante la
evacuacion del Titanic.

## 13. Conclusion final - Quienes tenian mas probabilidades de sobrevivir

A partir del analisis realizado, podemos concluir que los factores que
mas influyeron en la supervivencia de los pasajeros del Titanic fueron:

**1. Sexo:** Las mujeres tuvieron una tasa de supervivencia significativamente
mayor que los hombres, debido a la politica de evacuacion que priorizaba
mujeres y ninos.

**2. Clase del tiquete:** Los pasajeros de primera clase tuvieron mayor

**3. Edad:** Los ninos menores de 10 anos tuvieron mayor tasa de supervivencia
que los adultos, especialmente que los hombres adultos.

**4. Puerto de embarque:** Los pasajeros embarcados en Cherbourg tuvieron
mayor tasa de supervivencia, lo que se explica por la mayor proporcion
de pasajeros de primera clase en ese puerto.

**5. Grupo familiar:** Los pasajeros con familias pequenas tuvieron una
tasa de supervivencia ligeramente mayor que los que viajaban solos o
con familias grandes.

En conclusion, el perfil del pasajero con mas probabilidades de sobrevivir
era una mujer de primera clase, embarcada en Cherbourg, viajando con
una familia pequena. Por el contrario, el perfil con menos probabilidades
