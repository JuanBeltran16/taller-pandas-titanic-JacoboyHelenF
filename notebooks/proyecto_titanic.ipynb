## 8. Preguntas concretas sobre los pasajeros

### 8b. Cuantos pasajeros sobrevivieron y cuantos murieron

```python
# Conteo de sobrevivientes y fallecidos usando df_train
sobrevivieron = df_train[df_train['Survived'] == 1].shape[0]
murieron = df_train[df_train['Survived'] == 0].shape[0]
total = df_train.shape[0]

print("=== Supervivencia de pasajeros ===")
print(f"Sobrevivieron: {sobrevivieron} pasajeros ({sobrevivieron/total*100:.2f}%)")
print(f"Murieron:      {murieron} pasajeros ({murieron/total*100:.2f}%)")
print(f"Total:         {total} pasajeros")
```

### Observacion

De los 891 pasajeros del dataset train, menos de la mitad sobrevivio
al naufragio. La tasa de mortalidad fue significativamente alta, lo que
refleja las condiciones del desastre y la insuficiencia de botes salvavidas
para todos los pasajeros a bordo.

### 8d. Cuantos pasajeros viajaron con un familiar a bordo

```python
# Pasajeros que viajaron con al menos un familiar
con_familiar = df_train[df_train['Familiares'] > 0].shape[0]
sin_familiar = df_train[df_train['Familiares'] == 0].shape[0]
total = df_train.shape[0]

print("=== Pasajeros con y sin familiares ===")
print(f"Con familiar:  {con_familiar} pasajeros ({con_familiar/total*100:.2f}%)")
print(f"Sin familiar:  {sin_familiar} pasajeros ({sin_familiar/total*100:.2f}%)")
print(f"Total:         {total} pasajeros")
```

### Observacion

Una parte significativa de los pasajeros viajaba acompanada de al menos
un familiar a bordo. Esto es relevante porque mas adelante analizaremos
si viajar acompanado influyo en las probabilidades de supervivencia.

### 9b. Sobrevivieron mas ninos que hombres adultos proporcionalmente

Creemos que si, los ninos tuvieron mayor tasa de supervivencia que los
hombres adultos porque en situaciones de emergencia se prioriza evacuar
primero a los ninos.

```python
# Filtramos ninos y hombres adultos del dataset train
ninos = df_train[df_train['Age'] < 18]
hombres_adultos = df_train[(df_train['Age'] >= 18) & (df_train['Sex'] == 'male')]

tasa_ninos = ninos['Survived'].mean() * 100
tasa_hombres = hombres_adultos['Survived'].mean() * 100

print("=== Supervivencia: ninos vs hombres adultos ===")
print(f"Ninos (menores de 18):")
print(f"  Total:         {len(ninos)}")
print(f"  Sobrevivieron: {ninos['Survived'].sum()}")
print(f"  Tasa:          {tasa_ninos:.2f}%")

print(f"\nHombres adultos (18 o mas):")
print(f"  Total:         {len(hombres_adultos)}")
print(f"  Sobrevivieron: {hombres_adultos['Survived'].sum()}")
print(f"  Tasa:          {tasa_hombres:.2f}%")
```

### Observacion

Los datos confirman que los ninos tuvieron una tasa de supervivencia
significativamente mayor que los hombres adultos, lo que es consistente
con la politica de evacuacion que priorizaba mujeres y ninos.

### 9d. Los sobrevivientes partieron mas de Southampton que de otros puertos proporcionalmente

Creemos que no necesariamente, ya que aunque Southampton tenia mas pasajeros
en total, eso no implica que tuvieran mayor tasa de supervivencia. La clase
del tiquete y el sexo probablemente influyeron mas que el puerto de embarque.

```python
# Tasa de supervivencia por puerto de embarque
supervivencia_puerto = df_train.groupby('Embarked')['Survived'].agg(['sum', 'count', 'mean'])
supervivencia_puerto.columns = ['Sobrevivieron', 'Total', 'Tasa']
supervivencia_puerto['Tasa'] = (supervivencia_puerto['Tasa'] * 100).round(2)

print("=== Supervivencia por puerto de embarque ===")
print(supervivencia_puerto.to_string())
```

### Observacion

Los datos muestran que los pasajeros embarcados en Cherbourg tuvieron
una tasa de supervivencia mayor que los de Southampton y Queenstown.
Esto posiblemente se debe a que Cherbourg tenia una mayor proporcion
de pasajeros de primera clase, lo que influyo en sus probabilidades
de supervivencia.

## 10. Grupos familiares y distribucion de cabinas

```python
# Clasificacion de pasajeros en grupos familiares
def clasificar_familia(familiares):
    if familiares == 0:
        return 'Grupo 1: Sin familiares'
    elif familiares <= 3:
        return 'Grupo 2: Familia pequena'
    else:
        return 'Grupo 3: Familia grande'

df_train['grupo_familiar'] = df_train['Familiares'].apply(clasificar_familia)
df_test['grupo_familiar'] = df_test['Familiares'].apply(clasificar_familia)

# Distribucion de cabinas por grupo familiar
df_train['tiene_cabina'] = df_train['Cabin'].notna()

distribucion = df_train.groupby('grupo_familiar')['tiene_cabina'].agg(['sum', 'count', 'mean'])
distribucion.columns = ['Con cabina', 'Total', 'Proporcion']
distribucion['Proporcion'] = (distribucion['Proporcion'] * 100).round(2)

print("=== Distribucion de cabinas por grupo familiar ===")
print(distribucion.to_string())
```

### Observacion

Los pasajeros sin familiares a bordo representan el grupo mas grande.
Los pasajeros con cabina asignada son mayoritariamente de primera clase,
independientemente del grupo familiar. Las familias grandes tienen muy
poca presencia de cabinas asignadas, lo que sugiere que viajaban
principalmente en tercera clase.
